```python
filepath = 'test/IMG_1784.jpg'

threshold_values = [40, 80, 120, 160]
blur_values = [5, 15, 25, 35, 45]
```

```python
%cd ..
import image_utils
%pwd
plt.rcParams['figure.dpi'] = 150
```

```python
image = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception('Error, file \'%s\' not found.' % filepath)

for blur in blur_values:
    blurred_image = cv2.medianBlur(image, blur)
    image_utils.show_image(blurred_image)
    
    for bw_threshold in threshold_values:
        (thresh, monochrome_image) = cv2.threshold(blurred_image, bw_threshold, 1, cv2.THRESH_BINARY_INV)
        image_utils.show_image(monochrome_image, title='Blur: %s - Thresh: %s' % (blur, bw_threshold))
```
